```python
class Retangulo:
    def __init__(self, largura, comprimento):
        self.largura = largura #atributo publico
        self._comprimento = comprimento #atributo protegido
        self.__area = self.calcular_area() #atributo privado

    def calcular_area(self):
        return self.largura * self._comprimento

    def get_area(self):
        return self.__area

    def set_valores(self, largura, comprimento):
        self.largura = largura
        self._comprimento =  comprimento
        self.__area = self.calcular_area()

retangulo = Retangulo(5, 10)   

# Acesso aos atributos:
print(f"Largura: {retangulo.largura}") #Atributo público
print(f"Comprimento: {retangulo._comprimento}") #Atributo protegido

#Acesso à área usando o método de leitura
print(f"Área: {retangulo.get_area()}")

# Modificação dos valores usando o método set_valores
retangulo.set_valores(7, 12)

# Novo acesso à área usando o método de leitura
print(f"Nova Área: {retangulo.get_area()}")
```

# 02
#### Crie uma classe chamada ContaBancária que tenha os seguintes atributos:
#### Um atributo público chamado saldo.
#### Um atributo protegido chamado _titular.
#### Um atributo privado chamado __numero_conta.
#### Implemente métodos para definir e obter os valores de saldo, _titular e __numero_conta. Garanta que o saldo só possa ser modificado por métodos específicos de depósito e saque, e que o número da conta só possa ser acessado internamente.

```python
class ContaBancaria:
    def __init__(self, titular, numero_conta, saldo=0):
        self._titular = titular #Atribut
        self.__numero_conta = numero_conta #Atributo
        self.__saldo = saldo

    def get_saldo(self):
        return self.__saldo
    
    def depositar(self, valor):
        if valor > 0:
            self.__saldo += valor
            print(f"Depósito de R${valor} realizado. Novo saldo: R${self.__saldo}")
        else:
            print("Valor de saque inválido ou saldo insuficiente.")

    def sacar(self, valor):
        if valor > 0 and valor <= self.__saldo:
            self.__saldo -= valor
            print(f"Saque de R${valor} realizado. Novo saldo: R${self.__saldo}")
        else:
            print("Valor de saque inválido ou saldo insuficiente.")

    def get_titular(self):
        return self._titular

    def get_numero_conta(self):
        return self.__numero_conta

# Exemplo de uso:
conta = ContaBancaria(titular="João", numero_conta='123456', saldo=1000)

# Acessando os atributos
print(f'Titular: {conta.get_titular()}')
print(f'Número da Conta: {conta.get_numero_conta()}')
print(f"Saldo: R${conta.get_saldo()}")

#Realizando operações
conta.depositar(500)
conta.sacar(200)
```

# 3
#### Crie uma classe chamada Cafeteira que simula o funcionamento de uma cafeteira. Ela deve ter os seguintes atributos:
#### Um atributo público chamado água.
#### Um atributo protegido chamado _café.
#### Um atributo privado chamado __copos_preparados.
#### Implemente métodos para encher a cafeteira com água, adicionar café, preparar uma xícara de café e contar quantas xícaras já foram preparadas. Certifique-se de que a água e o café só possam ser acessados internamente pela classe e que o número de copos preparados seja apenas lido, não modificado diretamente. Garanta que, ao preparar uma xícara de café, a quantidade de água e café seja reduzida adequadamente.

```python
class Cafeteria:
    def __init__(self):
        self.agua = 0
        self._cafe = 0
        self.__copos_preparados = 0

    def encher_agua(self, quantidade):
        self.agua += quantidade
    
    def adicionar_cafe(self, quantidade):
        self._cafe += quantidade

    def contar_copos_preparados(self):
        return self.__copos_preparados

    def preparar_cafe(self):
        if self.agua >= 200 and self._cafe >= 10:
            self.agua -= 200
            self._cafe -= 10
            self.__copos_preparados += 1
            print("Café preparado!")
        else:
            print("Não há água ou café suficientes para preparar uma xícara.")

# Exemplo de uso da classe Cafeteira
cafeteira = Cafeteria()

cafeteira.encher_agua(1000)  # enche a cafeteira com 1000 ml de água
cafeteira.adicionar_cafe(100)  # adiciona 100 g de café

cafeteira.preparar_cafe()  # tenta preparar uma xícara de café
print("Copos preparados:", cafeteira.contar_copos_preparados())  # imprime o número de copos preparados

cafeteira.preparar_cafe()  # tenta preparar outra xícara de café
print("Copos preparados:", cafeteira.contar_copos_preparados())  # imprime o número atualizado de copos preparados

```
